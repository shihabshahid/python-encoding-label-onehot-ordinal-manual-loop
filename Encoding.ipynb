# Label Encoder Manual

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
df1.head()
```

```python
df1['sex']=le.fit_transform(df1['sex'])
df1['smoker']=le.fit_transform(df1['smoker'])
df1['region']=le.fit_transform(df1['region'])
```

```python
df1.head()
```

# Label Encoder Loop

```python
df2.head()
```

```python
for c in df2.columns:
    if is_numeric_dtype(df2[c]):
        continue
    df2[c]=le.fit_transform(df2[c])
```

```python
df2.head()
```

# One Hot Encoder Manual

```python
df3.head()
```

```python
one = pd.get_dummies(df3['sex'], drop_first=True, prefix='dummy')
df3.drop(df3[['sex']], axis=1, inplace=True)
df3 = pd.concat([df3, one], axis=1)
one = pd.get_dummies(df3['smoker'], drop_first=True, prefix='dummy')
df3.drop(df3[['smoker']], axis=1, inplace=True)
df3 = pd.concat([df3, one], axis=1)
one = pd.get_dummies(df3['region'], drop_first=True, prefix='dummy')
df3.drop(df3[['region']], axis=1, inplace=True)
df3 = pd.concat([df3, one], axis=1)
```

```python
df3.head()
```

# One Hot Encoder Loop

```python
df4.head()
```

```python
for c in df4.columns:
    if is_numeric_dtype(df4[c]):
        continue
    one = pd.get_dummies(df4[c], drop_first=True, prefix='dummy')
    df4.drop(df4[[c]], axis=1, inplace=True)
    df4 = pd.concat([df4, one], axis=1)
```
